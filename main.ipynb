```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# baca gambar (BGR)
img = cv2.imread('img-1.jpeg')

# ubah ke grayscale agar sharpening maksimal
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.figure(figsize=(5,5))
plt.imshow(gray, cmap='gray')
plt.title("Gambar Asli (Grayscale)")
plt.axis('off')
```

```python
# kernel sharpening
kernel = np.array([[0, -1, 0],
                   [-1, 5, -1],
                   [0, -1, 0]])

# lakukan sharpening
sharpened = cv2.filter2D(gray, -1, kernel)

plt.figure(figsize=(5,5))
plt.imshow(sharpened, cmap='gray')
plt.title("Hasil Sharpening")
plt.axis('off')
```

```python
plt.figure(figsize=(12,6))

plt.subplot(1,2,1)
plt.imshow(gray, cmap='gray')
plt.title("Asli")
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(sharpened, cmap='gray')
plt.title("Setelah Sharpening")
plt.axis('off')
```
